############ Limpar dados ############




```python
# single [] to value is deprecated
df_agrupados = file.groupby('Data')[['Óbitos','Recuperados','Ativos' ]].sum().reset_index()
df_agrupados.sort_values(by='Ativos', ascending=False)
# Agrupa lista descendente por maior numero de confirmados
```

```python
temp = file.groupby('Data')[['Óbitos','Ativos','Recuperados']].sum().reset_index()
#temp = file.groupby('Data')[['Confirmados','Notificações','Óbitos']].sum().reset_index()
```

```python
temp = temp.melt(id_vars="Data", value_vars=['Óbitos','Ativos','Recuperados'],
                 var_name='Case', value_name='Count')
#temp = temp.melt(id_vars="Data", value_vars=['Confirmados','Notificações','Óbitos'],var_name='Case', value_name='Count')
```

```python
temp.head(20)
temp.tail(20)
```

```python
# Habilita modo offline
from plotly.offline import plot, iplot, init_notebook_mode
init_notebook_mode(connected=True)

# Definindo o renderizador:
import plotly.io as pio
pio.renderers
pio.renderers.default = "colab"
```

```python
# Cores
ativos = '#21bf73'
mortes = '#ff2e63'
recuperados = '#fe9801'
```

```python

fig = px.area(temp, 
              x="Data", 
              y="Count", 
              color='Case', 
              height=600,
              title='Indice de recuperação dos casos',
              color_discrete_sequence = [ativos,mortes, recuperados])

fig.update_layout(xaxis_rangeslider_visible=True)
fig.show();
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=file.Data, y=file.Confirmados, # fill='tonexty' ,
                         name='Total de Casos Registrados', fill='tozeroy', mode='lines', line_color='indigo',
                        line = dict(color='rgb(22,26,28)')))
fig.add_trace(go.Scatter(x=file.Data, y=file.Recuperados,  name='Casos Recuperados', fill='tozeroy',
                        line = dict(color='#2CA02C', simplify=True,width=4)))
fig.add_trace(go.Scatter(x=file.Data, y=file.Ativos,  name='Casos Ainda Ativos',
                        line = dict(color='Red', simplify=True,width=4)))# fill down to xaxis
#fig.add_trace(go.Scatter(x=[1, 2, 3, 4], y=[3, 5, 1, 7], fill='tonexty')) # fill to trace0 y

fig.update_layout(legend=dict(x=0,y=1.0))
#fig.update_layout(xaxis_rangeslider_visible=True)

fig.show();
   
```

```python
# Média móvel dos ultimos 5 dias
rol_mean = file['Novos Casos'].rolling(7, win_type='triang', min_periods=1, center = True).mean()
rol_mean = pd.DataFrame(data = rol_mean)
#rol_mean = round(rol_mean).astype(int)
rol_mean.tail()
```

```python
fig = go.Figure(data = [go.Bar(x =file['Data'], y = file['Novos Casos'], name = 'Novos casos')], layout_title_text = 'Novos casos (por dia)')
fig.update_traces(marker_color = 'rgb(228,26,28)')
fig.add_trace(go.Scatter(x=file['Data'],y =  rol_mean['Novos Casos'],name= 'Média Móvel', mode='lines',line = dict(color='blue')))
fig.update_layout(legend=dict(x=0,y=1.0))
fig.show();
```
